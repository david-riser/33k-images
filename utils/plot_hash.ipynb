```python
import json 
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
with open('hashes_study.json') as inp:
    data= json.load(inp)
    
print(data)
```

```python
def extract_data(data, study_name, var_name):
    return [entry[var_name] for entry in data[study_name]]
```

```python
if not os.path.exists('figures'):
    os.mkdir('figures')
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,1,1)
plt.plot(
    extract_data(data, 'sample_size', 'sample_size'),
    extract_data(data, 'sample_size', 'hash_search_time'),
    marker='o',
    color='red',
    label='Hash Search'
)
plt.grid(alpha=0.2)
plt.xlabel('Sample Size')
plt.ylabel('Time (s)')
plt.legend(frameon=False)

plt.subplot(2,1,2)
plt.plot(
    extract_data(data, 'sample_size', 'sample_size'),
    extract_data(data, 'sample_size', 'diff_search_time'),
    marker='o',
    color='blue',
    label='Difference Search'
)
plt.grid(alpha=0.2)
plt.xlabel('Sample Size')
plt.ylabel('Time (s)')
plt.legend(frameon=False)
plt.savefig('figures/sample_size.png', bbox_inches='tight', dpi=100)
```

```python
image_sizes = extract_data(data, 'image_size', 'image_size')
print(image_sizes)

hash_times = extract_data(data, 'image_size', 'hash_search_time')
diff_times = extract_data(data, 'image_size', 'diff_search_time')
```

```python
plt.figure(figsize=(12,8))

plt.barh(np.arange(len(image_sizes)), hash_times, 0.4, edgecolor='k', color='purple', label='Hash Search')
plt.barh([0.5 + x for x in np.arange(len(image_sizes))], diff_times, 0.4, edgecolor='k', label='Difference Search',
        tick_label=[str(ims) for ims in image_sizes])
plt.xscale('log')
plt.grid(alpha=0.2)
plt.legend(frameon=False)
plt.xlabel('Time (s)')
plt.title('Image Size Scaling')
plt.savefig('figures/image_size.png', bbox_inches='tight', dpi=100)
```
